```python
from bs4 import BeautifulSoup
import requests
import csv
import time

with open('Age_winpct.csv', 'w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(["Team", "Average Age", "2023-11-01", '2023-12-01', '2024-01-01', '2024-02-01', '2024-03-01', '2024-04-01'])
    for team in range(1, 31):
        cumulative_age = 0
        players = 0
        player_row = 2
        age_page = requests.get(f"https://cleaningtheglass.com/stats/team/{team}#tab-offensive_overview")
        soup = BeautifulSoup(age_page.text, "html.parser")
        team_list = soup.find('select', id='team-name-select-desktop')
        current_team = team_list.find('option', {'value': f'{team}'}) ### Gets FULL team name from CLeaningTHeGlass, ex: 'Atlanta Hawks'
        table = soup.find('table')
        while players != 9:
            row = table.find_all('tr')[player_row]
            age = row.find_all('td')[1]
            cumulative_age += float(age.text.strip())
            players += 1
            player_row += 1
            time.sleep(3)
        teams_dict = {'Atlanta Hawks': 'ATL', 'Boston Celtics': 'BOS', 'Brooklyn Nets': 'BRK', 'Charlotte Hornets': 'CHO', 'Chicago Bulls': 'CHI', 'Cleveland Cavaliers': 'CLE', 'Dallas Mavericks': 'DAL', 'Denver Nuggets': 'DEN', 'Detroit Pistons': 'DET', 'Golden State Warriors': 'GSW', 'Houston Rockets': 'HOU', 'Indiana Pacers': 'IND', 'Los Angeles Clippers': 'LAC', 'Los Angeles Lakers': 'LAL', 'Memphis Grizzlies': 'MEM', 'Miami Heat': 'MIA', 'Milwaukee Bucks': 'MIL', 'Minnesota Timberwolves': 'MIN', 'New Orleans Pelicans': 'NOP', 'New York Knicks': 'NYK', 'Oklahoma City Thunder': 'OKC', 'Orlando Magic': 'ORL', 'Philadelphia 76ers': 'PHI', 'Phoenix Suns': 'PHO', 'Portland Trail Blazers': 'POR', 'Sacramento Kings': 'SAC', 'San Antonio Spurs': 'SAS', 'Toronto Raptors': 'TOR', 'Utah Jazz': 'UTA', 'Washington Wizards': 'WAS'}
        eastern_teams = ['Atlanta Hawks', 'Boston Celtics', 'Brooklyn Nets', 'Charlotte Hornets', 'Chicago Bulls', 'Cleveland Cavaliers', 'Detroit Pistons', 'Indiana Pacers', 'Miami Heat', 'Milwaukee Bucks', 'New York Knicks', 'Orlando Magic', 'Philadelphia 76ers', 'Toronto Raptors', 'Washington Wizards']
        western_teams = ['Dallas Mavericks', 'Denver Nuggets', 'Golden State Warriors', 'Houston Rockets', 'Los Angeles Clippers', 'Los Angeles Lakers', 'Memphis Grizzlies', 'Minnesota Timberwolves', 'New Orleans Pelicans', 'Oklahoma City Thunder', 'Phoenix Suns', 'Portland Trail Blazers', 'Sacramento Kings', 'San Antonio Spurs', 'Utah Jazz']
        first_of_month_indices = [12, 42, 71, 104, 129, 162]
        eastern_standings = requests.get('https://www.basketball-reference.com/leagues/NBA_2024_standings_by_date_eastern_conference.html')
        western_standings = requests.get('https://www.basketball-reference.com/leagues/NBA_2024_standings_by_date_western_conference.html')
        east_soup = BeautifulSoup(eastern_standings.text, 'html.parser')
        west_soup = BeautifulSoup(western_standings.text, 'html.parser')
        east_table = east_soup.find('table', id='standings_by_date')
        west_table = west_soup.find('table', id='standings_by_date')
        percentages = []
        for number in first_of_month_indices:
            if current_team.text in eastern_teams:
                row = east_table.find_all('tr')[number]
            else:
                row = west_table.find_all('tr')[number]
            specific_team = row.find('td', attrs = {'class': f'left {teams_dict[current_team.text.strip()]}'})
            record = specific_team.find('span', attrs = {'class':'note'})
            clean = record.text.replace('-', '.').replace('(', '').replace(')', '')
            num1, num2 = map(int, clean.split('.'))
            pct = round(float(num1 / (num1+num2)), 3)
            percentages.append(pct)
            time.sleep(3)
        time.sleep(10)
        writer.writerow([current_team.text.strip(), round((cumulative_age / 9), 2), percentages[0], percentages[1], percentages[2], percentages[3], percentages[4], percentages[5]])
        print(f'Added {current_team.text.strip()} data to file.')
file.close()
print('All done!')
```
